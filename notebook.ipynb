```python
# notebook.ipynb

import pandas as pd
from meteostat import Stations, Daily
from weather_analysis.data_analis.data_processing import WeatherDataProcessor, save_to_excel

# Настройки
start = pd.Timestamp('2023-01-01')
end = pd.Timestamp('2023-12-31')

# Получение данных о погоде
stations = Stations()
station = stations.nearby(55.7558, 37.6173).fetch(1)
station_id = station.index[0]

data = Daily(station_id, start, end)
data = data.fetch()

# Преобразование данных
data = data[['tavg']]
data.columns = ['temperature']
data.dropna(inplace=True)

# Анализ данных
processor = WeatherDataProcessor(data)
processed_data = processor.process_data()

# Сохранение результата в Excel
save_to_excel(processed_data, 'weather_analysis.xlsx')

# Выводим промежуточные результаты
processed_data.head()
```
